```python
import numpy as np
from sklearn.metrics import mean_squared_error

import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import average_precision_score


data = pd.read_csv('Ground Water .csv')



#data = data.fillna(data.mean())
# Fill NaN values only in numeric columns with median
numeric_columns = data.select_dtypes(include='number').columns
data[numeric_columns] = data[numeric_columns].fillna(data[numeric_columns].median())


# Select only numeric columns for quantile calculations
numeric_data = data.select_dtypes(include=[np.number])
data[numeric_data.columns] = numeric_data.clip(lower=numeric_data.quantile(0.01), upper=numeric_data.quantile(0.99), axis=1)

#data = data.clip(lower=data.quantile(0.01), upper=data.quantile(0.99), axis=1)

data = pd.get_dummies(data)

scaler = StandardScaler()
data[data.select_dtypes(include=['float64']).columns] = scaler.fit_transform(data.select_dtypes(include=['float64']))

data = data.drop(data.columns[[1, 2]], axis=1)
X = data.iloc[:,:-3]
y = data.iloc[:, -1]

#print("\nX: ",X)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)

#print("\nx: ")
#print("\nx:",X_train)
#print(X_train.shape())

from sklearn.model_selection import KFold
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

model = LinearRegression()

kf = KFold(n_splits=5, shuffle=True)


mse_scores = []
for train_index, test_index in kf.split(X_train):
    # Split the data into training and testing sets for this fold
    X_train_fold, X_test_fold = X_train.iloc[train_index], X_train.iloc[test_index]
    y_train_fold, y_test_fold = y_train.iloc[train_index], y_train.iloc[test_index]
    # Train the model on the training set and test it on the testing set
    model.fit(X_train_fold, y_train_fold)
    y_pred = model.predict(X_test_fold)
    mse = mean_squared_error(y_test_fold, y_pred)
    mse_scores.append(mse)

# Compute the average mean squared error across all folds
avg_mse = sum(mse_scores) / len(mse_scores)
print("\navgmse: ",avg_mse)

Y_train=np.unique(y_train_fold)
#y_train=y_train.reshape(-1,1)
print("\ny_train: ",y_train)
print("\ny_train shape: ",y_train.shape)

#neural Network
input_layer_size = X_train.shape[1]
print("\ninput_layer_shape: ",X_train.shape[1])
import numpy as np


# Neural Network with weight-setting from outside
class NeuralNetwork:
    def __init__(self, input_layer_size, hidden_layer_size, output_layer_size):
        self.input_layer_size = input_layer_size
        self.hidden_layer_size = hidden_layer_size
        self.output_layer_size = output_layer_size

    def set_weights(self, weights):
        # Extract weights and biases from flat vector
        idx = 0
        self.W1 = weights[idx:idx + self.input_layer_size * self.hidden_layer_size].reshape(self.input_layer_size, self.hidden_layer_size)
        idx += self.input_layer_size * self.hidden_layer_size

        self.b1 = weights[idx:idx + self.hidden_layer_size].reshape(1, self.hidden_layer_size)
        idx += self.hidden_layer_size

        self.W2 = weights[idx:idx + self.hidden_layer_size * self.output_layer_size].reshape(self.hidden_layer_size, self.output_layer_size)
        idx += self.hidden_layer_size * self.output_layer_size

        self.b2 = weights[idx:idx + self.output_layer_size].reshape(1, self.output_layer_size)

    def sigmoid(self, x):
        x=np.asarray(x,dtype=float)
        return 1 / (1 + np.exp(-x))

    def forward_propagation(self, X):
        self.Z1 = np.dot(X, self.W1) + self.b1
        self.A1 = self.sigmoid(self.Z1)
        self.Z2 = np.dot(self.A1, self.W2) + self.b2
        self.A2 = self.sigmoid(self.Z2)
        return self.A2

    def loss(self, y_pred, y_true):
        y_true = y_true.reshape(-1, 1)
        return mean_squared_error(y_true, y_pred)

    def predict(self, X):
        output = self.forward_propagation(X)
        predictions = (output >= 0.01).astype(int)
        for i, pred in enumerate(predictions):
            if pred == 1:
                print(f"Sample {i+1}: Water is harmful")
            else:
                print(f"Sample {i+1}: Water is not harmful")
        return predictions


class WOA:
    def __init__(self, obj_func, n_whale, spiral_constant, n_iter, lb, ub):
        self.obj_func = obj_func
        self.n_whale = n_whale
        self.spiral_constant = spiral_constant
        self.n_iter = n_iter
        self.lb = lb
        self.ub = ub

    def initialize(self):
        self.positions = np.random.uniform(self.lb, self.ub, (self.n_whale, len(self.lb)))
        self.fitness = np.array([self.obj_func(w) for w in self.positions])
        self.best_idx = np.argmin(self.fitness)
        self.best_position = self.positions[self.best_idx].copy()
        self.best_fitness = self.fitness[self.best_idx]

    def update_position(self, t):
        a = 2 - t * (2 / self.n_iter)
        for i in range(self.n_whale):
            r1, r2 = np.random.rand(), np.random.rand()
            A = 2 * a * r1 - a
            C = 2 * r2
            p = np.random.rand()

            if p < 0.5:
                if abs(A) >= 1:
                    rand_idx = np.random.randint(0, self.n_whale)
                    D = abs(C * self.positions[rand_idx] - self.positions[i])
                    self.positions[i] = self.positions[rand_idx] - A * D
                else:
                    D = abs(C * self.best_position - self.positions[i])
                    self.positions[i] = self.best_position - A * D
            else:
                D = abs(self.best_position - self.positions[i])
                l = np.random.uniform(-1, 1)
                self.positions[i] = D * np.exp(self.spiral_constant * l) * np.cos(2 * np.pi * l) + self.best_position

            # Clip within bounds
            self.positions[i] = np.clip(self.positions[i], self.lb, self.ub)

    def run(self):
        self.initialize()
        for t in range(self.n_iter):
            self.update_position(t)
            self.fitness = np.array([self.obj_func(w) for w in self.positions])
            current_best_idx = np.argmin(self.fitness)
            if self.fitness[current_best_idx] < self.best_fitness:
                self.best_fitness = self.fitness[current_best_idx]
                self.best_position = self.positions[current_best_idx].copy()
        return self.best_position, self.best_fitness



input_size = X_train.shape[1]
hidden_size = 10
output_size = 1

# Total weights = W1 + b1 + W2 + b2
total_weights = input_size * hidden_size + hidden_size + hidden_size * output_size + output_size
lb = -1 * np.ones(total_weights)
ub = 1 * np.ones(total_weights)

def fitness_function(weights):
    model = NeuralNetwork(input_size, hidden_size, output_size)
    model.set_weights(weights)
    output = model.forward_propagation(X_train)
    return model.loss(output, y_train.values.reshape(-1, 1))

# Run WOA
woa = WOA(fitness_function, n_whale=20, spiral_constant=1, n_iter=30, lb=lb, ub=ub)
best_weights, best_fitness = woa.run()

# Evaluate final model
model = NeuralNetwork(input_size, hidden_size, output_size)
model.set_weights(best_weights)

predictions = model.predict(X_test)
test_accuracy = ((predictions == y_test.values.reshape(-1, 1)).mean()) * 100
print(f"\nTest Accuracy: {test_accuracy:.2f}%")
print(f"Final MSE (Loss): {best_fitness:.4f}")


# (optional) Plot if needed
plt.plot([fitness_function(w) for w in woa.positions])
plt.title("Fitness per Whale at Final Iteration")
plt.xlabel("Whale")
plt.ylabel("Loss")
plt.show()

```
